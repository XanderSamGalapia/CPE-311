```python
filepath = '/content/chocolate.csv'
import numpy as np
import pandas as pd
```

```python
choc = pd.read_csv(filepath)
choc
```

```python
choc.columns
```

```python
#Renaming
choc.rename(columns = {'company_location': 'Company'}, inplace = True)
choc
```

```python
#Renaming/shorten the name of country of bean origin
choc.rename(columns = {'country_of_bean_origin': 'Bean_Origin'}, inplace = True)
choc
```

```python
#Removing the ref
choc.drop(['ref'], axis = 1 , inplace=True)
choc
```

```python
##All chocolate ratings below 2
rate2 = choc[choc['rating'] <= 2].copy()
rate2
```

```python
##All chocolate ratings above 3
rate3 = choc[choc['rating'] >= 3].copy()
rate3
```
