```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
salaries
```

##Mean

```python
##MEAN
def mean(salaries):
  return sum(salaries)/len(salaries)

salary_mean = sum(salaries)/len(salaries)
print("Salaries Mean: ",salary_mean)
```

##Median

```python
##Median
sorted_salaries = sorted(salaries)
n = len(salaries)

if n % 2 == True:
  salary_median = sorted_salaries[n // 2]
else:
  salary_median = (sorted_salaries[n // 2 - 1] + sorted_salaries[n // 2])/2

print("Salary Median: ", salary_median)
```

##Mode

```python
##Mode
def calculate_mode(salaries):
    counts = {}
    for num in salaries:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1

    max_count = max(counts.values())
    mode = [num for num, count in counts.items() if count == max_count]
    return mode

mode = calculate_mode(salaries)
print("Mode:", mode)
```

##Sample Variance

```python
##Sample Variance
def sam_var(salaries):
  n = len(salaries)
  if n < 2:
    return None

  mean = sum(salaries)/n
  squared_diff_sum = sum((x - mean) ** 2 for x in salaries)
  sample_variance = squared_diff_sum / (n - 1)

  return sample_variance

sample_variance = sam_var(salaries)
print("Sample Variance:", sample_variance)
```

##Sample Standard Deviation

```python
##Sample Standard Deviation
def sam_dev(salaries):
    sample_variance = sam_var(salaries)
    if sample_variance is None:
        return None
    return sample_variance ** 0.5

sam_dev = sam_dev(salaries)
print("Sample Standard Deviation:", sam_dev)
```

##Coefficient of variation Interquartile range

```python
##Range
def Coeff_var(salaries):
    mean = sum(salaries) / len(salaries)
    SD = (sum((x - mean) ** 2 for x in salaries) / len(salaries)) ** 0.5
    return SD / mean

def Intrq_range(data):
  if not data:
    return None
  sorted_data = sorted(data)
  midpoint = len(sorted_data) // 2
  q1 = sorted_data[midpoint // 2] if len(sorted_data) % 2 else (sorted_data[midpoint - 1] + sorted_data[midpoint]) / 2
  q3 = sorted_data[midpoint + (midpoint % 2)] if len(sorted_data) % 2 else (sorted_data[midpoint + midpoint // 2 - 1] + sorted_data[midpoint + midpoint // 2]) / 2
  return q3 - q1


salaries_cv = Coeff_var(salaries)
salaries_iqr = Intrq_range(salaries)

print("Coefficient: ", salaries_cv)
print("Interquartile:" , salaries_iqr)


```

##Quartile coefficient of dispersion

```python
##Quartile coefficient of dispersion
def QCOED(salaries):
    if not salaries:
        return None
    salaries_sorted = sorted(salaries)
    q1_index = len(salaries_sorted) // 4
    q3_index = 3 * len(salaries_sorted) // 4
    q1 = salaries_sorted[q1_index]
    q3 = salaries_sorted[q3_index]
    return (q3 - q1) / (q3 + q1)

qcd = QCOED(salaries)
print("Quartile Coefficient of Dispersion:", qcd)
```
